# El comando ```gh```.

*Github* permite realizar diversas operaciones con *GitHub* desde una terminal mediante el comando [```gh```](https://cli.github.com/).

El comando ```gh``` puede ser instalado y ejecutado en *Windows*, *MacOS X* y *GNU/Linux*. El el siguiente enlace contiene las instrucciones de instalación:

* https://cli.github.com/manual/installation

El manual de ```gh``` puede ser consultado des
de:

* https://cli.github.com/manual/

```
gh <ambito> <argumentos>
```

* Donde:

* ```<ambito>``` es un ámbito *scope* autorizado de *Github*.
* ```<argumentos>``` es una serie de argumentos del comando.

## Actividad: *Login* de ```gh``` mediante *SSH* y un *PAT*.

**Nota:** Se da por sentado que el comando ```gh``` ya fue instalado.

* S ejecuta el comando ```gh auth login```.


* Selecciona *SSH* como el portocolo preferido para ```gh```.


* Se indica, la localización de la llave pública del usuario. En el caso de que esta *notebook* se ejecute desde la *VM* , la ruta es ```/home/oi/.ssh/id_rsa.pub```.



* Se define el nombre para la llave *SSH*. En este caso se define el nonmbre por defecto.


* Se selecciona la opción de autenticarse mendiante un *PAT*.

* En este caso, se generará un *PAT* con los ámbitos ```repo```, ```read:org``` y ```admin:public_key``` habilitados.

* Se ingresa el *PAT*.

* Si los datos son correctos, el comando ```gh```estará habilitado y las operaciones en el sistema estarán ligadas a la cuenta del usuario de *GitHub* que posee el *PAT*.


* La siguiente celda enlistará los reositorios de la organización ```Cloudevel```.

````bash
gh repo ls Cloudevel
````

* La siguiente celda mostrará el readme del repositorio ```PythonistaMX/py101```.

````bash
gh repo view PythonistaMX/py101
````

````bash
gh org list
````

* Si se ejecuta el comando ```gh``` sin argumentos, traerá la ```man page```.

````bash
gh
````

* Para desligarse del usuario de *GitHub* se usa el comando ```gh auth logout```.

````bash
gh auth logout
````
