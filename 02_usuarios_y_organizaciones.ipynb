## Cuentas de usuario.

Para poder acceder a los servicios de *Github*, es necesario crear una cuenta de usuario.

Registrarse como usuario de *GitHub* sólo requiere de una cuenta de correo electrónico válida y por defecto se tendrá acceso a la suscripción gratuita para personas. En caso de optar por otro tipo de cuenta, es necesario contar con algún método de pago válido.

Las cuentas de usuario están ligadas a una persona, por lo que es posible usarla para proyectos personales y también para colaborar en otros proyectos de *GitHub*.

Cada usuario de *GitHub* debe registrar un nombre único y un medio de autenticación, el cual puede ser una contraseña o mediante un esquema de múltiples factores de autenticación (*MFA*).

* https://docs.github.com/es/get-started/onboarding/getting-started-with-your-github-account



### Acceso a los recursos de *GitHub*.

Un usuario puede acceder a los recursos de *GitHub* de las siguiente formas:

* [El sitio web](https://github.com/).
* [*Github Desktop*](https://desktop.github.com/)
* Un *IDE*, como por ejemplo *Visual Studio Code* o *Eclipse*.
* La *CLI* usando los comandos ```git``` y [```gh```](https://cli.github.com/manual/).
* Las *API* [*REST*](https://docs.github.com/es/rest) o [*GraphQL*](https://docs.github.com/es/graphql) de *GitHub*.

### El perfil del usuario.

El perfil del usuario permite acceder a información pública del usuario, tal como:

* Los repoostiorios, poryectos y paquetes públicos en los que colasbora o le pertenecen.
* Su actividad.
* Sus datos de contacto.
* Las organizaciones en las que colabora.
* Algunos reconocimientos.
* Los proyectos o repositorios que ha marcado con una estrella.



### La configuración del usuario.

La configuración del usuario (*settings*) es el espacio en le que el usuario puede establecer diversos aspectos de sus actividades en *Github*.

Para acceder as la configuración se selecciona la opción Settings del menú desplegable doned se encuentra el avatar del usuario en la esquina superior derecha.

La página de configuración permite gestionar diversos aspectos del usuario, tanto conrespecto a su información como a temas como facturación y seguridad.



El menú lateral contiene diversas configuraciones agrupadas en:

* Preferencias del usuario (no tiene encabezado).
* Acceso (*Access*).
* Código, planeación y automatización (*Code, panning and automation).
* Seguridad (*Security*).
* Integraciones (*Integrations).
* Archivos (*Archives*).
* Configuración de desarrollador (*Developer settings*).


A continuación se enumeran algunas de las opciones más relevantes: 

* ```Access | Billing and plans``` permite definir.
  * El tipo de plan de suscripción.
  * Los límites de gastos.
  * Las formas de pago.
  
  

* ```Access | Passwords and authentication``` permite configurar la forma en la que el usuarios se pueden autenticar en el sitio.


https://docs.github.com/es/authentication



* ```Acess | SSH an GPG Keys``` permite ingresar las llaves públicas *SSH* y *PGP* del usuario, de tal forma que sirva como método de autuenticación para la gestión de repositorios locales.



* ```Integraciones | Aplicaciones``` permite configurar.
 * Las [aplicaciones de *GitHub*](https://docs.github.com/en/apps) instaladas y autorizadas.
 
 
 
 * Las aplicaciones externas que utilizan la autenticación *OAuth* de *GitHub* como método de autenticación.




* ```Developer Settings```. Donde es posible configurar.
  * Las *GitHub apps* creadas por el usuario.
  * Las *Aplicaciones* registradas por el usuario.
  * Los *tokens* personales de acceso (*PAT*).



## Organizaciones.

Los usuarios pueden crear, formar parte y colaborar en organizaciones.

La gestión de las organizaciones que posee un usuario, se configurar desde ```Access | Organizations```.



Las organizaciones representan un tipo de identidad de *GitHub* que permite tener diversos recursos para empresas o grupos organizados.





* Las organizaciones cuentan con su propio espacio de recursos, esquemas de suscripción y configuraciones.
