## *Github Desktop*.

*GitHub Desktop* es una aplicación de escritorio para *MacOS X* y *Windows* que permite gestionar diversas funcionalidades de *GitHub* sin necesidad de acceder al sitio.

https://desktop.github.com/



## Pages.

https://pages.github.com/

GitHub Pages es un servicio de alojamiento de sitio estático que toma archivos *HTML*, *CSS* y *JavaScript* directamente desde un repositorio en *GitHub*. Opcionalmente ejecuta los archivos a través de un proceso de complilación y publica un sitio web.

Es posible hospedar el sitio en el dominio ```github.io``` de *GitHub* o un dominio personalizado propio.

Las organizaciones que usan GitHub Enterprise Cloud también pueden publicar sitios de forma privada si administran el control de acceso del sitio. 

La siguiente liga muestra algunos ejemplos de *GitHub Pages*. 

* https://github.com/collections/github-pages-examples

El siguente enlace contiene la documentación de *Pages*.

* https://docs.github.com/es/pages/getting-started-with-github-pages/about-github-pages

## Projects.

Un proyecto es una hoja de cálculo adaptable que se integra con las incidencias y *pull requests* en *GitHub* para facilitar la  planeación y seguimiento del trabajo de forma eficaz. Es posible crear y personalizar varias vistas mediante el filtrado, la ordenación y la agrupación de las incidencias y *pull requests*, visualizar el trabajo con gráficos configurables y agregar campos personalizados para realizar el seguimiento de metadatos específicos del equipo. En lugar de aplicar una metodología específica, un proyecto proporciona características flexibles que puedes personalizar en función de las necesidades y los procesos del equipo.

Los proyectos se crean a partir de las incidencias y los *pull requests* que se igeneran, lo que crea referencias directas entre el proyecto y el trabajo. La información se sincroniza automáticamente con el proyecto a medida que se realizan cambios, y se actualizan las vistas y los gráficos. Esta integración es bidireccional, por lo que,  la información sobre un *pull requests* o una incidencia del proyecto cambia, el *pull requests* o la incidencia reflejará la información. Es posible ampliar aún más esta integración, agrupar el proyecto por usuario asignado y realizar cambios en la asignación de incidencias.

* https://docs.github.com/es/issues/planning-and-tracking-with-projects/learning-about-projects/about-projects

## *Dependabot*.

*Dependabot* es un sistema automatizado que verifica las dependecias de un repositorio en *Github*.

Las *Dependabot security updates* facilitan la correción de dependencias vulnerables en un repositorio. Al habilitar esta característica, cuando se levante una alerta del *Dependabot* para una dependencia vulnerable en la gráfica de dependencias del repositorio, Dependabot intentará arreglarla automáticamente. *GitHub* podría enviar alertas a los repositorios.

*Dependabot* verifica si es posible actualizar la dependencia vulnerable a una versión arreglada sin irrumpir en la gráfica de dependencias para el repositorio. Posteriormente, *Dependabot* levanta un *pull request* para actualizar la dependencia a la versión mínima que incluye el parche y los enlaces a la solicitud de cambios para la alerta del *Dependabot*, o reporta un error en la alerta.

https://docs.github.com/es/code-security/dependabot/dependabot-security-updates/about-dependabot-security-updates

## *Codespaces*.

https://github.com/features/codespaces

Un *codespace* es un ambiente de desarrollo que se hospeda en la nube. Esposible personalziar un proyecto para *GitHub Codespaces* mediante un *commit* de los archivos de configuración en el repositorio (lo que se conoce a menudo como configuración como código). Esta acción crea una configuración de *codespace* repetible para todos los usuarios del proyecto.

Cada *codespace* se hospeda en *GitHub* en un contenedor *Docker*, el cual se ejecuta en una máquina virtual. Es posilbe elegirr entre una selección de tipos de máquina virtual que tienen desde 2 núcleos, 8 GB de RAM y 32 GB de almacenamiento hasta 32 núcleos, 64 GB de RAM y 128 GB de almacenamiento.

De forma predeterminada, los *codespaces* se crean a partir de una imagen de *Ubuntu* que incluye una selección de lenguajes y herramientas populares, pero es posible usar una imagen basada en cualquier distribución de *GNU/Linux* y configurarla. Independientemente del sistema operativo local, el *codespace* se ejecutará en un entorno de *GNU/Linux*. *Windows* y *MacOS* no son sistemas operativos compatibles con el contenedor remoto.

https://docs.github.com/es/codespaces
