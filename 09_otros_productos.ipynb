https://pages.github.com/

GitHub Pages es un servicio de alojamiento de sitio esttico que toma archivos *HTML*, *CSS* y *JavaScript* directamente desde un repositorio en *GitHub*. Opcionalmente ejecuta los archivos a travs de un proceso de complilacin y publica un sitio web.

## Projects.

Un proyecto es una hoja de clculo adaptable que se integra con las incidencias y *pull requests* en *GitHub* para facilitar la  planeacin y seguimiento del trabajo de forma eficaz. Es posible crear y personalizar varias vistas mediante el filtrado, la ordenacin y la agrupacin de las incidencias y *pull requests*, visualizar el trabajo con grficos configurables y agregar campos personalizados para realizar el seguimiento de metadatos especficos del equipo. En lugar de aplicar una metodologa especfica, un proyecto proporciona caractersticas flexibles que puedes personalizar en funcin de las necesidades y los procesos del equipo.

Los proyectos se crean a partir de las incidencias y los *pull requests* que se igeneran, lo que crea referencias directas entre el proyecto y el trabajo. La informacin se sincroniza automticamente con el proyecto a medida que se realizan cambios, y se actualizan las vistas y los grficos. Esta integracin es bidireccional, por lo que,  la informacin sobre un *pull requests* o una incidencia del proyecto cambia, el *pull requests* o la incidencia reflejar la informacin. Es posible ampliar an ms esta integracin, agrupar el proyecto por usuario asignado y realizar cambios en la asignacin de incidencias.

* https://docs.github.com/es/issues/planning-and-tracking-with-projects/learning-about-projects/about-projects

## *Dependabot*.

*Dependabot* es un sistema automatizado que verifica las dependecias de un repositorio en *Github*.

Las *Dependabot security updates* facilitan la correcin de dependencias vulnerables en un repositorio. Al habilitar esta caracterstica, cuando se levante una alerta del *Dependabot* para una dependencia vulnerable en la grfica de dependencias del repositorio, Dependabot intentar arreglarla automticamente. *GitHub* podra enviar alertas a los repositorios.

*Dependabot* verifica si es posible actualizar la dependencia vulnerable a una versin arreglada sin irrumpir en la grfica de dependencias para el repositorio. Posteriormente, *Dependabot* levanta un *pull request* para actualizar la dependencia a la versin mnima que incluye el parche y los enlaces a la solicitud de cambios para la alerta del *Dependabot*, o reporta un error en la alerta.

https://docs.github.com/es/code-security/dependabot/dependabot-security-updates/about-dependabot-security-updates

## *Codespaces*.

https://github.com/features/codespaces

Un *codespace* es un ambiente de desarrollo que se hospeda en la nube. Esposible personalziar un proyecto para *GitHub Codespaces* mediante un *commit* de los archivos de configuracin en el repositorio (lo que se conoce a menudo como configuracin como cdigo). Esta accin crea una configuracin de *codespace* repetible para todos los usuarios del proyecto.

Cada *codespace* se hospeda en *GitHub* en un contenedor *Docker*, el cual se ejecuta en una mquina virtual. Es posilbe elegirr entre una seleccin de tipos de mquina virtual que tienen desde 2 ncleos, 8 GB de RAM y 32 GB de almacenamiento hasta 32 ncleos, 64 GB de RAM y 128 GB de almacenamiento.

De forma predeterminada, los *codespaces* se crean a partir de una imagen de *Ubuntu* que incluye una seleccin de lenguajes y herramientas populares, pero es posible usar una imagen basada en cualquier distribucin de *GNU/Linux* y configurarla. Independientemente del sistema operativo local, el *codespace* se ejecutar en un entorno de *GNU/Linux*. *Windows* y *MacOS* no son sistemas operativos compatibles con el contenedor remoto.

https://docs.github.com/es/codespaces
