# 03 Repositorios.

El componente base de *GitHub* son los repositorios remotos de *Git*. Tanto los usuarios como las organizaciones pueden gestionar repositorios piblicos y privados.




La documentacin de los repositorios est disponibe en:

https://docs.github.com/es/repositories/creating-and-managing-repositories/about-repositories

## Uso de los repositorios desde el sitio web de *Github*.

Un repositorio de *GitHub* puede ser gestionado directamente desde la pgina de *GitHub* de manera similar a un repositorio local de *Git*.

La interfaz grfica permite acceder y extender las fucnionalidades de *Git*.

### Caractersticas de los repositorios de *GitHub*.

* Seguridad extendida. El acceso a los repositorios de *GitHub* desde un entorno local requieren de autenticacin mediante un *PAT*, una llave *SSH* o mediante la aplicacin  ```gh```.
* Los repositorios pueden ser creados [en blanco](https://docs.github.com/en/repositories/creating-and-managing-repositories/creating-a-new-repository) o estar basados [en plantillas (templates)](https://docs.github.com/en/repositories/creating-and-managing-repositories/creating-a-repository-from-a-template).
* Muchas funcionalidades de *Git* son funcionalidades implementadas en el repositorio remoto. 
* Posibilidad de proteger y [gestionar las ramas](https://docs.github.com/en/repositories/configuring-branches-and-merges-in-your-repository/managing-protected-branches).
* Definir acceso a colaboradores.
* Acceso a *Forks* y *Pull requests*.
* Permite realizar operaciones de *CI/CD*.
* Puede ligarse a proyectos.
* Permite la interaccin mediante discusiones e incidentes (*issues*).
* Permite crear wikis.
* Cuenta con la opcin de usar [archivos grandes](https://docs.github.com/en/repositories/working-with-files/managing-large-files/about-large-files-on-github).
* Cuenta con funciones de [seguridad avanzadas](https://docs.github.com/en/repositories/managing-your-repositorys-settings-and-features/enabling-features-for-your-repository/managing-security-and-analysis-settings-for-your-repository) para el cdigo.
